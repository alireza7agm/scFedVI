```python
import numpy as np
import pandas as pd
import harmonypy as hp
import seaborn as sns
import matplotlib.pyplot as plt
```

## scVI

```python
a, b, c, d = 0, 0, 0, 0
for i in range(5):
    print(f'CLIENT {i}\n')
    print('----------------------')
    cell_types = pd.read_csv(f'../5/{i}_celltype.csv')
    batches = pd.read_csv(f'../5/{i}_tech.csv')

    meta_data = pd.DataFrame({
        'celltype': cell_types.iloc[:, 0],
        'tech': batches.iloc[:, 0]
    })
    
    data = np.load(f'../5/latent_scVI_{i}.npy')
    res = hp.compute_lisi(data, meta_data, ['celltype', 'tech'])
    meta_data['lisi_cell_type'] = res[:, 0]
    meta_data['lisi_batch'] = res[:, 1]
    
    
    df = meta_data
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(df['lisi_cell_type'], kde=True)
    plt.title('Distribution of LISI Scores for Cell Types')

    plt.subplot(1, 2, 2)
    sns.histplot(df['lisi_batch'], kde=True)
    plt.title('Distribution of LISI Scores for Batches')

    plt.tight_layout()
    plt.show()


    print("Average LISI score for Cell Types:", df['lisi_cell_type'].mean())
    a+=df['lisi_cell_type'].mean()
    print("STD LISI score for Cell Types:", df['lisi_cell_type'].std())
    b+=df['lisi_cell_type'].std()
    print("Average LISI score for Batches:", df['lisi_batch'].mean())
    c+=df['lisi_batch'].mean()
    print("STD LISI score for Batches:", df['lisi_batch'].std())
    d+=df['lisi_batch'].std()
    
    batch_stats = df.groupby('tech').agg({
    'lisi_cell_type': ['mean', 'median', 'std'],
    'lisi_batch': ['mean', 'median', 'std']
    })
    print(batch_stats)


print(a/5)
print(b/5)
print(c/5)
print(d/5)
```

```python
from scipy import stats

for i in range(2):
    print(f'CLIENT {i}\n')
    print('----------------------')
    cell_types = pd.read_csv(f'{i}_celltype.csv')
    batches = pd.read_csv(f'{i}_tech.csv')

    meta_data = pd.DataFrame({
        'celltype': cell_types.iloc[:, 0],
        'tech': batches.iloc[:, 0]
    })
    
    data = np.load(f'latent_scVI_{i}.npy')
    res_scVI = hp.compute_lisi(data, meta_data, ['celltype', 'tech'])
    meta_data['lisi_cell_type'] = res_scVI[:, 0]
    meta_data['lisi_batch'] = res_scVI[:, 1]
    
    data = np.load(f'latent_scFedVI_{i}.npy')
    res_scFedVI = hp.compute_lisi(data, meta_data, ['celltype', 'tech'])
    meta_data['lisi_cell_type'] = res_scFedVI[:, 0]
    meta_data['lisi_batch'] = res_scFedVI[:, 1]
    
    # cell types
    t_statistic, p_value = stats.ttest_ind(res_scVI[:, 0], res_scFedVI[:, 0])
    print("T-test p-value:", p_value)
    
    # batches
    t_statistic, p_value = stats.ttest_ind(res_scVI[:, 1], res_scFedVI[:, 1])
    print("T-test p-value:", p_value)
```
